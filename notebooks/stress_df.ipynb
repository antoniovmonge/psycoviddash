```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('../raw_data/cleaned_data_040321.csv',index_col=0)
```

```python
df['PSS10_avg'].min()
```

```python
df['PSS10_avg'].max()
```

```python
plt.bar(df['Dem_gender'].unique(), df.groupby('Dem_gender')['PSS10_avg'].mean())
```

```python
df_2 = pd.DataFrame(df.groupby(['Country', 'Dem_gender'])['PSS10_avg'].mean()).reset_index()
df_3 = df_2[df_2['Dem_gender'] != 'Other/would rather not say']
df_3
```

```python
plt.figure(figsize=(16,6))
ax = sns.barplot(x='Country', y='PSS10_avg',data=df_3,hue='Dem_gender')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
```
